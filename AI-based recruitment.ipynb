```
# AI-Based Recruitment Screening Lab
# ---------------------------------
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 2: Create synthetic candidate dataset
data = {
    "Name": [f"Candidate_{i}" for i in range(1, 21)],
    "Skills": [
        "Python, Data Analysis, Machine Learning",
        "Java, Spring Boot, Microservices",
        "Excel, HR Analytics, Employee Engagement",
        "SQL, Tableau, Data Visualization",
        "Python, Pandas, NumPy",
        "Project Management, Agile, Scrum",
        "Digital Marketing, SEO, Content Writing",
        "C++, Embedded Systems, IoT",
        "Finance, Accounting, Risk Management",
        "Cloud Computing, AWS, DevOps",
        "Python, NLP, Deep Learning",
        "JavaScript, React, Node.js",
        "HR Policies, Recruitment, Training",
        "Business Strategy, Consulting, Analytics",
        "Python, Statistics, Data Science",
        "Cybersecurity, Networking, Ethical Hacking",
        "Healthcare Management, Research, Analytics",
        "AI, Machine Learning, TensorFlow",
        "Marketing, Sales, CRM",
        "Python, SQL, Power BI"
    ],
    "Experience": np.random.randint(1, 10, 20),  # years of experience
    "Education": [
        "MBA", "B.Tech", "MBA", "B.Sc", "M.Sc", "MBA", "BBA", "B.Tech", "MBA", "B.Tech",
        "M.Tech", "B.Sc", "MBA", "MBA", "M.Sc", "B.Tech", "MBA", "M.Tech", "BBA", "M.Sc"
    ],
    "Certifications": [
        "Data Science", "Java Developer", "HR Analytics", "Tableau", "Python", "PMP",
        "Digital Marketing", "IoT", "Finance", "AWS", "Deep Learning", "Web Development",
        "HR", "Consulting", "Statistics", "Cybersecurity", "Healthcare", "AI", "Sales", "Power BI"
    ]
}

df = pd.DataFrame(data)

# Step 3: Define job description
job_description = """
We are looking for a Data Scientist with strong skills in Python, Machine Learning,
Data Analysis, and experience with tools like Pandas, NumPy, and Power BI.
An MBA or M.Sc in a relevant field is preferred. Certifications in Data Science or AI are a plus.
"""

# Step 4: TF-IDF + Cosine Similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["Skills"].tolist() + [job_description])
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()
df["Similarity"] = cosine_sim

# Step 5: Weighted Scoring Model
def weighted_score(row):
    score = 0
    # Similarity weight
    score += row["Similarity"] * 0.5
    # Experience weight
    score += (row["Experience"] / 10) * 0.2
    # Education weight
    if row["Education"] in ["MBA", "M.Sc", "M.Tech"]:
        score += 0.2
    # Certification weight
    if row["Certifications"] in ["Data Science", "AI", "Deep Learning", "Python"]:
        score += 0.1
    return score

df["Final_Score"] = df.apply(weighted_score, axis=1)

# Step 6: Shortlist top candidates
shortlist = df.sort_values(by="Final_Score", ascending=False).head(5)

# Step 7: Display results
print("Top Candidates Shortlist:\n")
print(shortlist[["Name", "Skills", "Experience", "Education", "Certifications", "Final_Score"]])
```

```
# Step 5: Convert Text into Numerical Form Using TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer

# Combine candidate skills and job description into one list
documents = df["skills"].tolist() + [job_description]

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Display the shape of the TF-IDF matrix
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)

# Optional: Show feature names (words extracted)
print("\nSample Features (words):", vectorizer.get_feature_names_out()[:20])
```

```
# This cell combines the job description and all candidate skills into one list.
# Then it converts the text into a TF-IDF matrix.
# The matrix is a numerical representation of the text.

documents = [job_description] + df["skills"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```
# Step 6: Calculate Cosine Similarity

from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity between job description (last row) and all candidate skill rows
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

# Add similarity scores to the dataset
df["Cosine_Similarity"] = cosine_sim

# Display top 5 candidates based on similarity
print("Top 5 Candidates by Cosine Similarity:\n")
print(df[["candidate_id", "name", "skills", "Cosine_Similarity"]].sort_values(by="Cosine_Similarity", ascending=False).head())
```

```
# This cell calculates cosine similarity between:
# 1. the job description vector
# 2. each candidate skill vector
#
# The result is stored as a similarity score for each candidate.

job_vector = tfidf_matrix[0:1]
candidate_vectors = tfidf_matrix[1:]

similarity_scores = cosine_similarity(job_vector, candidate_vectors).flatten()

df["similarity_score"] = similarity_scores

print("Cosine similarity scores calculated successfully.")
df[["candidate_id", "name", "skills", "similarity_score"]]
```

```
# Step 7: Normalize Experience and Certifications

from sklearn.preprocessing import MinMaxScaler

# ✅ Check if required columns exist
required_cols = ['experience_years', 'certifications']
for col in required_cols:
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in dataset")

# ✅ Handle missing values
df['experience_years'] = df['experience_years'].fillna(0)
df['certifications'] = df['certifications'].fillna(0)

# ✅ Initialize scaler
scaler = MinMaxScaler()

# ✅ Normalize the columns (0 to 1 scale)
df[['Experience_norm', 'Certifications_norm']] = scaler.fit_transform(
    df[['experience_years', 'certifications']]
)

# ✅ Display result
print(df[['experience_years', 'Experience_norm', 'certifications', 'Certifications_norm']].head())
```

```
# This cell normalizes experience and certification values.
# Normalization means converting values into a common range.
# Here we divide each value by the maximum value in its column.

df["experience_score"] = df["experience_years"] / df["experience_years"].max()
df["certification_score"] = df["certifications"] / df["certifications"].max()

df[["name", "experience_years", "experience_score", "certifications", "certification_score"]]
```

```
# Step 8: Assign Education Scores

# ✅ Check if column exists
if 'education' not in df.columns:
    raise ValueError("Column 'education' not found in dataset")

# ✅ Handle missing values
df['education'] = df['education'].fillna('').astype(str)

# ✅ Define education scoring (custom mapping)
education_score_map = {
    "B.Tech": 3,
    "B.Sc": 3,
    "BBA": 3,
    "BCA": 3,
    "B.Com": 3,
    "MBA": 4,
    "MBA HR": 4,
    "M.Sc": 4,
    "M.Tech": 4,
    "MCA": 4,
    "M.Stat": 4
}

# ✅ Convert education into numerical score
df['Education_score'] = df['education'].map(education_score_map)

# ✅ Handle unknown categories (if any)
df['Education_score'] = df['Education_score'].fillna(0)

# ✅ Display result
print(df[['education', 'Education_score']].head())
```

```
# This cell maps education categories into numerical scores.
# Higher values are assigned to higher or more specialized qualifications.
# This is just a teaching example, not a universal hiring rule.

education_map = {
    "B.Com": 0.50,
    "BBA": 0.55,
    "BCA": 0.55,
    "B.Sc": 0.60,
    "B.Tech": 0.70,
    "MBA": 0.75,
    "MBA HR": 0.80,
    "MCA": 0.80,
    "M.Sc": 0.80,
    "M.Tech": 0.90,
    "M.Stat": 0.90
}

df["education_score"] = df["education"].map(education_map)

df[["name", "education", "education_score"]]
```

```
# Step 9: Create a Weighted Final Score

from sklearn.metrics.pairwise import cosine_similarity

# ✅ Compute cosine similarity (job description vs resumes)
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()

# ✅ Add similarity score to dataframe
df['Similarity_score'] = cosine_sim

# ✅ Normalize similarity score (optional but recommended)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Similarity_score'] = scaler.fit_transform(df[['Similarity_score']])

# ✅ Normalize Education score (if not already normalized)
df['Education_norm'] = df['Education_score'] / df['Education_score'].max()

# ✅ Final Weighted Score Calculation
df['Final_score'] = (
    0.50 * df['Similarity_score'] +
    0.25 * df['Experience_norm'] +
    0.15 * df['Education_norm'] +
    0.10 * df['Certifications_norm']
)

# ✅ Sort candidates by Final Score (highest first)
df = df.sort_values(by='Final_score', ascending=False)

# ✅ Display top candidates
print(df[['Similarity_score', 'Experience_norm', 'Education_norm', 'Certifications_norm', 'Final_score']].head())
```

```
# This cell creates the final weighted score for each candidate.
# It combines text similarity, experience, education, and certifications.

df["final_score"] = (
    0.50 * df["similarity_score"] +
    0.25 * df["experience_score"] +
    0.15 * df["education_score"] +
    0.10 * df["certification_score"]
)

df[["name", "similarity_score", "experience_score", "education_score", "certification_score", "final_score"]]
```

```
# Step 10: Rank Candidates

# ✅ Sort candidates by Final Score (highest first)
df_sorted = df.sort_values(by='Final_score', ascending=False)

# ✅ Reset index for clean ranking
df_sorted = df_sorted.reset_index(drop=True)

# ✅ Assign rank (1 = best candidate)
df_sorted['Rank'] = df_sorted.index + 1

# ✅ Display ranked candidates
print(df_sorted[['Rank', 'Final_score']].head(10))
```

```
# This cell sorts candidates in descending order based on the final score.
# The highest-scoring candidate appears first.

ranked_df = df.sort_values(by="final_score", ascending=False).reset_index(drop=True)

ranked_df[["candidate_id", "name", "skills", "experience_years", "education", "certifications", "final_score"]]
```

```
# Step 11: Generate Top 5 Shortlisted Candidates

# ✅ Select top 5 candidates from ranked dataframe
top_5_candidates = df_sorted.head(5)

# ✅ Display shortlisted candidates
print("Top 5 Shortlisted Candidates:\n")
print(top_5_candidates[['Rank', 'Final_score']])
```

```
# This cell extracts the top 5 candidates from the ranked dataset.
# These candidates are considered the strongest matches according to the model.

top_5 = ranked_df.head(5)

top_5[["candidate_id", "name", "skills", "experience_years", "education", "certifications", "final_score"]]

```

```
# Step 12: Add HR Interpretation Labels

# ✅ Function to assign labels based on Final Score
def assign_label(score):
    if score >= 0.75:
        return "Strong Match"
    elif score >= 0.50:
        return "Moderate Match"
    else:
        return "Low Match"

# ✅ Apply function to create new column
df_sorted['HR_Recommendation'] = df_sorted['Final_score'].apply(assign_label)

# ✅ Display results
print(df_sorted[['Rank', 'Final_score', 'HR_Recommendation']].head(10))
```

```
# This cell converts final numeric scores into simple decision labels.
# This makes the result easier to interpret from an HR point of view.

def label_candidate(score):
    if score >= 0.65:
        return "Strong Match"
    elif score >= 0.50:
        return "Moderate Match"
    else:
        return "Low Match"

ranked_df["recommendation"] = ranked_df["final_score"].apply(label_candidate)

ranked_df[["candidate_id", "name", "final_score", "recommendation"]]

```

```
# Step 13: Final Recruitment Output

# ✅ Select important columns for HR review
columns_to_display = [
    'Rank',
    'name' if 'name' in df_sorted.columns else None,
    'Final_score',
    'HR_Recommendation',
    'experience_years',
    'education',
    'certifications'
]

# ✅ Remove None values (if 'Name' column doesn't exist)
columns_to_display = [col for col in columns_to_display if col is not None]

# ✅ Create final output table
final_output = df_sorted[columns_to_display]

# ✅ Display final recruitment table
print("Final Recruitment Output:\n")
print(final_output.head(10))
```
